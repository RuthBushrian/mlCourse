# K-Means Clustering

import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np
```

### creating dataset 
First of all, we need to create the data. We use the synthetic data of "blobs" from the Scikit-learn package


```python
centers = 5
X_train, true_labels = make_blobs(n_samples=1000, centers=centers, random_state=67)
```

Lets plot the dataset

```python
axis = sns.scatterplot(x=[X[0] for X in X_train], y=[X[1] for X in X_train], hue=true_labels, palette="deep", legend=None)
plt.xlabel("x")
plt.ylabel("y")
```

### K-Means Clustering

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have to specify the ***number of clusters k*** we want the data to be grouped into.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(5)
kmeans.fit(X_train)
# kmeans.labels_
```

lets plot the classification of kmeans

```python
axis = sns.scatterplot(x=[X[0] for X in X_train], y=[X[1] for X in X_train], hue=kmeans.labels_, palette="deep")
plt.xlabel("x")
plt.ylabel("y")
```

lets predict some points

```python
kmeans.predict([[-7.5, -7.5]])
```

what are the centers of the cluters?

```python
kmeans.cluster_centers_
```

### Find k

Specify a range of k values to try

```python
wcss = []

k_values = range(1, 11)
```

Calculate WCSS for each k

```python
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

Plot the WCSS values to find the "elbow"

```python
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()
```
